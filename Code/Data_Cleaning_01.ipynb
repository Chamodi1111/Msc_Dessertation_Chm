```python
#Import Libraries and modules
import pandas as pd
from scipy.stats import skew, kurtosis
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
#import matplotlib.pyplot as plt
#import numpy as np
#from sklearn.metrics import mean_absolute_error, r2_score
#from sklearn.metrics import accuracy_score, classification_report
#from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load the dataset
data = pd.read_csv("Transactions.csv")
```

```python
print("Skewness:", skew(data['meter_sale_price']))
print("Kurtosis:", kurtosis(data['meter_sale_price']))
```

```python
#Drop the unwanted columns
drop_columns = [
    'trans_group_ar', 'procedure_name_ar', 'property_type_ar', 'property_sub_type_ar',
    'property_usage_ar', 'reg_type_ar', 'area_name_ar', 'building_name_ar', 'project_name_ar',
    'master_project_ar', 'nearest_landmark_ar', 'nearest_metro_ar', 'nearest_mall_ar', 'rooms_ar', 'building_name_en', 'project_number', 'master_project_en', 'property_sub_type_id', 'property_sub_type_en', 'rent_value', 'meter_rent_price', 'project_name_en', 'no_of_parties_role_1', 'no_of_parties_role_2', 'no_of_parties_role_3' 
]
data.drop(columns=drop_columns, inplace=True, errors='ignore')
```

```python
pd.set_option('display.max_columns', None)
data.head(100)
```

```python
#checking null value counts
nan_per_column = data.isna().sum()
print("\nNaN values per column:")
print(nan_per_column)
```

```python
# Drop rows where 'rooms_en' has null values
data = data[data['rooms_en'].notna()]
```

```python
# Handle missing values in specific columns based on 'area_name_en'
specific_columns = ['nearest_landmark_en', 'nearest_metro_en', 'nearest_mall_en']

for col in specific_columns:
    for area in data['area_name_en'].unique():
        area_mask = data['area_name_en'] == area
        most_frequent_value = data.loc[area_mask, col].mode()
        if not most_frequent_value.empty:
            data.loc[area_mask & data[col].isna(), col] = most_frequent_value[0]

# If still missing values exist, use KNN Imputation
knn_columns = ['nearest_landmark_en', 'nearest_metro_en', 'nearest_mall_en']
ordinal_encoder = OrdinalEncoder()

# Encoding categorical columns to numeric for KNN
data[knn_columns] = data[knn_columns].fillna('Unknown')
data[knn_columns] = ordinal_encoder.fit_transform(data[knn_columns])

knn_imputer = KNNImputer(n_neighbors=5)
data[knn_columns] = knn_imputer.fit_transform(data[knn_columns])

# Decode numeric back to original labels
data[knn_columns] = ordinal_encoder.inverse_transform(data[knn_columns])
```

```python
# Handle missing values in 'actual_worth' based on 'meter_sale_price' and 'procedure_area'
if 'actual_worth' in data.columns and 'meter_sale_price' in data.columns and 'procedure_area' in data.columns:
    missing_worth_mask = data['actual_worth'].isna()
    data.loc[missing_worth_mask, 'actual_worth'] = (
        data.loc[missing_worth_mask, 'meter_sale_price'] * data.loc[missing_worth_mask, 'procedure_area']
    )
```

```python
most_frequent_date = data['instance_date'].mode()[0]
data['instance_date'].fillna(most_frequent_date, inplace=True)
```

```python
print(data.isnull().sum())
```

```python
# Drop rows with any remaining NaNs (from division)
data.dropna(inplace=True)
```
